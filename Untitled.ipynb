```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('nba-2022-UTC.csv')
```

```python
data
```

```python
data.info()
```

```python
df = data.copy()
```

```python
day_and_time = df['Date'].apply(lambda x: x.split(' '))
```

```python
day = []
time = []
for i in day_and_time:
    day.append(i[0])
    time.append(i[1])
df['Day'] = day
df['Time'] = time
```

```python
df['Result'] = df['Result'].astype(str) 
sep_result = df['Result'].apply(lambda x: x.split('-'))

home_team_result = []
away_team_result = []
for i in sep_result:
    if len(i) > 1:
        home_team_result.append(i[0])
        away_team_result.append(i[1])
    else:
        home_team_result.append('nan')
        away_team_result.append('nan')
df['Home_result'] = home_team_result
df['Away_result'] = away_team_result
```

```python
df_clean = df.drop(['Location','Match Number','Date','Result'], axis=1)
df_clean = df_clean[(df_clean['Home_result'] != 'nan') & (df_clean['Away_result'] != 'nan')]
# df_clean[['Home_result','Away_result']] = df_clean[['Home_result','Away_result']].astype(float)
```

```python
df_clean
```

```python
df_clean[['Home_result','Away_result']] = df_clean[['Home_result','Away_result']].astype(float)
```

```python
df_clean['Points'] = df_clean['Home_result'] + df_clean['Away_result']
```

```python
# neural network - regrission
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
df_clean = pd.get_dummies(df_clean,columns = ['Home Team','Away Team','Day','Time'])
```

```python
x_data = df_clean.drop(['Points','Home_result','Away_result'], axis=1)
y_data = df_clean['Points']
```

```python
scaler_1 = StandardScaler()
scaler_2 = MinMaxScaler()
scaler = scaler_2

x_data = scaler.fit_transform(x_data)
y_data = np.array(y_data)
y_data = y_data.reshape(-1,1)
y_data = scaler.fit_transform(y_data)
```

```python
x_data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=10)
```

```python
# Linear Regresssion
model_1 = LinearRegression()
model_1.fit(x_train,y_train)
y_pred = model_1.predict(x_test)
```

```python
score = r2_score(y_pred,y_test)
score
```

```python
# DecisionTreeRegressor
model_2 = DecisionTreeRegressor()
model_2.fit(x_train,y_train)
y_pred_tree = model_2.predict(x_test)
```

```python
score = mean_absolute_error(y_pred_tree,y_test)
score
```

```python
cross_val_score(model_2,x_test,y_test, cv=3)
```
